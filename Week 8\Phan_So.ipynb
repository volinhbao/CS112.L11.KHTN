#**PHÂN SỐ**

Cho 2 phân số đúng và tối giản $\frac{a}{b}$ , $\frac{c}{d}$  . Mỗi phép biến đổi là tăng $a$ và $b$ lên 1, sau đó giản ước phân số nhận được.

Hãy xác định sau bao nhiêu bước biến đổi từ phân số thứ nhất ban đầu nhận được phân số thứ 2 đã cho.

**Dữ liệu**: Vào từ thiết bị nhập chuẩn gồm 4 dòng, mỗi dòng chứa 1 số nguyên $a, b, c, d, 0 < a < b ≤ 105, 0 < c < d ≤ 105$, $a$ và $b$ nguyên tố cùng nhau, $c$ và $d$ nguyên tố cùng nhau  $ \frac{a}{b} \neq \frac{c}{d}$

**Kết quả**: Đưa ra thiết bị xuất chuẩn số 0 nếu không có cách biến đổi hoặc một số nguyên – số lượng phép biến đổi.

## Abstract
Với phép biến đổi phân số: $f(\frac{a}{b}) = \frac{a+1}{b+1}$,

Tính số bước biến đổi từ phân số $\frac{a}{b}$ thành $\frac{c}{d}$

## Pattern Recognition
Khi xét các điều kiện $a < b$ và $c < d$ ta nhận ra những đặc điểm sau:


 $a < b  <=> \frac{a}{b} < \frac{a+1}{b+1}$

# Algorithm

## Method 1: Brute-force ($O(n^2)$)
Just merely follow the steps metioned in the problem statement:

- Increment (add 1) $a$ and $b$
- Simplify new fraction $\frac{a}{b}$ 
- Compare new $\frac{a}{b}$ to $\frac{c}{d}$, if they are not equal then increment variable ```count```

Pseudo code


````
tu_1 = int(input().strip())
mau_1 = int(input().strip())
tu_2 = int(input().strip())
mau_2 = int(input().strip())
count = 0
# gcd = math.gcd(fractions[0],fractions[1])
while(tu_1!= tu_2 or mau_1 != mau_2):
    count += 1
    tu_1 += 1
    mau_1 += 1
    a = tu_1
    b = mau_1
    while (b!=0):
      temp = b
      b = a % b
      a = temp
    # print("a: {0}, b: {1}".format(a,b))
    # print(a)
    tu_1 //= a
    mau_1 //= a
    # print("tu: {0}, mau: {1}".format(fractions[0],fractions[1]))
    
print(count)
````
