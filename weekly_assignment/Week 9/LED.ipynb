#**Đèn LED**
Alice vừa hoàn thành một hợp đồng lắp bảng hiển thị số bằng đèn LED. Mỗi chữ số được hiển thị trong một khung chữ nhật với 7 ống đèn LED. Bằng cách bật các ống đèn LED thích hợp ta có thể hiển thị chữ số bất kỳ. Số đèn sáng càng nhiều thì việc hiển thị chữ số đó càng tốn năng lượng. Ví dụ, hiển thị chữ số 9 sẽ tốn năng lượng hơn hiển thị chữ số 7.

Hãy xác định tổng lớn nhất của các chữ số của số có thể bật sáng.

**Dữ liệu:** Vào từ thiết bị nhập chuẩn gồm một dòng chứa số nguyên **$n$** ($2 ≤ n ≤ 10^6$).

**Kết quả:** Đưa ra thiết bị xuất chuẩn một số nguyên – tổng lớn nhất đạt được.

##Pattern Recognition
Đây là bài toán có thể giải bằng mẹo toán học:
- Ta nhận thấy chữ số 7 là số có giá trị lớn nhất mà được tạo bởi số lượng đèn led nhỏ nhất (=3)
- Để được tổng lớn nhất của các chữ số thỏa mãn yêu cầu đề bài, ta sẽ cố gắng lấy nhiều chữ số 7 nhất có thể, nếu còn dư 1 bóng ta sẽ bớt 1 chữ số 7 để lấy 1 chữ số 4 (tạo bởi 4 bóng led). Nếu còn dư 2 bóng ta sẽ lấy thêm 1 chữ số 1.


## Algorithm
- Bước 1: Ta thực hiện phép chia lấy dư của $n$ với 3.
- Bước 2: Nếu phần dư bằng 1: lấy kết quả của phép chia lấy nguyên $n$ cho 3 trừ đi 1, sau đó nhân với 7 và cộng thêm 4. Đó là tổng lớn nhất thỏa mãn yêu cầu đề bài.
- Nếu phần dư bằng 2: lấy kết quả phép chia lấy nguyên $n$ cho 3 nhân với 7 và cộng thêm 1.
- Nếu $n$ chia hết cho 7: lấy kết quả phép chia lấy nguyên $n$ cho 3 nhân với 7.

Pseudo code
```
n <- input
left_led <- n mod 7
num_7 <- n div 7
if left_led == 2:
  res <- (num_7 - 1) * 7 + 4
  end if
if left_led == 1:
  res <- num_7 * 7 + 1
  end if
if left_led == 0:
  res <- num_7 * 7
  end if
```
