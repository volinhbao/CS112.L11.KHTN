## Vấn đề xác nhận điểm nằm giữa hai tia
Gọi góc $\alpha_3$ là góc tạo bởi đường nối điểm đang xét Y($x_1, y_1$) và X($x_0,y_0$) với truc hoành. Điểm Y nằm giữa 2 tia đi từ điểm X hợp với trục hoàng một góc $\alpha_1, \alpha_2$ nếu thỏa điều kiện sau:


> $\alpha_1 < \alpha_3 < \alpha_2$

> $\therefore tan(\alpha_1) < tan(\alpha_3) < tan(\alpha_2)$

> $\therefore \frac{a}{b} < \frac{y_1 - y_0}{x_1 - x_0} < \frac{c}{d}$

> $\therefore by_1 - ax_1 > by_0 - ax_0$ and $cx_1 - dy_1 > cx_0 - dy_0$

Nhìn vào công thức cuối, ta thấy cặp tọa độ điêm X($x_0,y_0$) và Y($x_1, y_1$) có thể được biểu diễn dưới dạng một con số $by_1 - ax_1$ hoặc $cx_1 - dy_1$ hoặc $by_0 - ax_0$ hoặc $cx_0 - dy_0$.




## Tiếp cận theo hướng đồ thị 
Từ các phân tích ở trên, một ý tưởng đơn giản có thể nghĩ đến là về việc xây dựng 1 đồ thị có hướng. Mỗi đỉnh của đồ thị là 1 điểm của đề bài. Mỗi cạnh có hướng $(u,v)$ biểu diễn 1 cặp đỉnh sao cho đỉnh $v$ nằm bên trong góc nhìn từ điểm $u$ (nằm giữa 2 tia xuất phát từ $u$). Tới đây bài toán trở thành bài tìm cum dài nhất trên đồ thị có hướng. Ta có thể giải bằng thuật toán duyệt sau DFS

Thuật toán có độ phức tạp về thời gian là O($n^2$). Chi phí lưu trữ là O($n^2$). Thuật toán này không phù hợp với kích thước của bài toán

## 1 hướng tiếp cận khác 
Một cách khác để giải bài toán đếm sao, ta chọn cách biểu diễn $cx - dy$, ứng với góc $\alpha_2$ và tiến hành sort giảm dần (độ phức tạp $O(nlog)$) tất cả tọa độ ngôi sao sau khi chuyển về dạng $cx_i - dy_i$

##Bài toán dãy con tăng dài nhất
Cho một mảng số nguyên: 
> **1** 4 **2 3** 1 0

dãy con tăng dài nhất là:

> 1 2 3

**Note:** 


*   Dãy con là tập hợp các số nguyên thuộc một mảng theo đúng trực tự của mảng đó, không nhất thiết phải liên tục
*   Dãy con tăng là dãy con có tính chất tăng dần


## Thuật giải bài toán dãy con tăng dài nhất 
### Thuật toán 1:
#### Ý tưởng
Sử dụng ý tưởng quy hoạch động, ta gọi $dp[i]$ là độ dài dãy con tăng dài nhất kết thúc tại vị trí $i$  
Công thức quy hoạch động  
- $dp[i]=max_{1 \le j \le i}(dp[j])\ +\ 1$  

Ta sử dụng 2 vòng lặp chồng nhau; vòng lặp thứ nhất duyệt i qua từng phần tử của mảng; vòng lặp thứ 2 duyệt j từ 1 đến i. Cập nhật $dp[i]$ từ các $dp[j]$ theo công thức quy hoạch động ở trên.  
Kết quả bài toán là $max_{1 \le i \le n}(dp[i])$
#### Đánh giá:
Thuật toán có độ phức tạp O($n^2$)

````
n = int(input("Nhập n "))
a = list(map(int,input("Nhập dãy số nguyên ").split()))
dp = []
ret = -10**12
#Duyệt từng cặp điểm i,j và cập nhật giá trị tại i theo các giá trị tại j
for i in range(n):
  x = 1
  for j in range(i):
    if (a[i]>a[j]):
      x = max(x,dp[j] + 1)
  dp.append(x)
  ret = max(ret,x)
print("Dộ dài dãy con tăng dài nhất là {}".format(ret))
````

### Thuật toán 2:
#### Ý tưởng
Sử dụng ý tưởng quy hoạch động kết hợp chặt nhị phân, ta gọi $dp[i]$ là độ dài dãy con tăng dài nhất kết thúc tại vị trí $i$, $lis[i]$ là giá trị nhỏ nhất trong mảng thỏa điều kiện dãy con tăng dài nhất kết thúc tại vị trí này có độ dài là $i$  
$lis[i]\ =\ min_{dp[j]=i}(a[j])$  
Bởi vì lis là 1 dãy tăng tuyến tính nên ta có thể sử dụng ý tưởng chặt nhị phân. Tại mỗi bước $i$, t tìm giá trị $j$ lớn nhất sao cho $lis[j] < a[i]$  
Sau đó cập nhật $dp[i]$  và $lis[dp[j]]$
- $dp[i]=j+1$
- $lis[dp[i]] = min(lis[dp[i],a[i])$

````
n = int(input("Nhập n "))
a = list(map(int,input("Nhập dãy số nguyên ").split()))
dp = []
lis = [-10**12]
for i in range(n):
  # Chặt nhị phân trên mảng lis
  low = 1
  high = len(lis) - 1
  pos = 1
  while low<=high:
    mid = (low+high)//2
    if a[i] > lis[mid]:
      pos = mid + 1
      low = mid + 1
    else:
      high = mid -1

  #Cập nhật mảng lis với kết quả vừa tính được tại phần tử thứ i
  if pos == len(lis):
    lis.append(pos)
  else:
    lis[pos] = a[i]
print("Dộ dài dãy con tăng dài nhất là {}".format(len(lis)-1))
````

# Pseudo Code
Ở đây sẽ trình bày pseudo code cho thuật giải O($nlog(n)$) theo hướng tiếp cận bài dãy con tăng dài nhất 
```
input -> n
input -> a,b,c,d
input -> point[1],point[2],...,point[n]

Sắp xếp mảng point theo thứ tự tăng dần c*point[0] - d*point[1]
Loop i <- 1 to n
  Chặt nhị phân j lớn nhất thỏa lis[j] < b*point[1] - a*point[0]
  dp[i] = j + 1
  lis[dp[i]] = min(lis[dp[i]], b*point[1] - a*point[0])

output <- len(lis)
```
# Execution cho bài toán đếm sao

````
n = int(input())
a,c = input().split()[:2]
a,b = a.split('/')
c,d = c.split('/')

a=int(a)
b=int(b)
c=int(c)
d=int(d)

points = []

for i in range(1,n+1):
	x,y = map(int,input().split()[:2])
	if b*y-a*x > 0 and d*y-c*x < 0:
		points.append((x,y))

def mykey(x):
	return (c*x[0] - d*x[1],a*x[0] - b*x[1])

points.sort(key=mykey)
lis = []
for point in points:
	low = 0
	high = len(lis) - 1
	ret = 0
	while low <= high:
		mid = (low + high)//2
		if lis[mid] < b*point[1] - a*point[0]:
			ret = mid + 1
			low = mid + 1
		else:
			high = mid - 1
	if ret == len(lis):
		lis.append(b*point[1] - a*point[0])
	else:
		lis[ret] = b*point[1] - a*point[0]
print(len(lis))
````
